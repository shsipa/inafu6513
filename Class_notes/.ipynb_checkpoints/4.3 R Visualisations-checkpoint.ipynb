# R Visualisations

## Basic visualisations

```r
b <- c(1,3,7,5,3,2)
names(b) <- c('jan','feb','mar','apr','may','jun')
barplot(b)
```

```r
b <- c(1,3,7,5,3,2)
f <- sqrt(b)
plot(b,f)
```

## Handing missing values

```r
b <- c(1,3,7,NA,3,2)
sum(b)
sum(b, na.rm=TRUE)
```

## Matrices (two-dimensional arrays)

```r
x <- matrix(0, 4, 3)
y <- matrix(1:12, 4, 3)
y[2,3] <- 15
y[,3]
y[2,2:3]
print(y[1,3])
z <- 1:15
dim(z) <- c(3,5)
print(z)
```

## Visualising Matrices
R comes with example matrices, including volcano.

```r
y <- matrix(1:12, 4, 3)
contour(y)
persp(y)
```

```r
contour(volcano)
persp(volcano)
```

## Getting statistics

```r
b <- c(1,3,7,5,3,2)
mean(b)
median(b)
sd(b)
```

## Adding to visualisations

```r
b <- c(1,3,7,5,3,2)
names(b) <- c('jan','feb','mar','apr','may','jun')
mean(b)
barplot(b)
abline(h=mean(b))
```
